```python
%matplotlib inline

import numpy as np
import pandas as pd

from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
test_ids = test.Id

train.head(5)
```

```python
train.corr()['Cover_Type']
```

```python
data = train.append(test)
data['Cover_Type'] = data['Cover_Type'].fillna(-1).astype(int)
data = data.drop(['Id'], axis=1)
```

```python
X_train = data[:len(train)]
X_test = data[len(train):]
```

```python
y_train = X_train.Cover_Type
X_train = X_train.drop(['Cover_Type'], axis=1)

X_train, X_cv, y_train, y_cv = train_test_split(X_train, y_train, test_size=0.15)
X_train
```

```python
X_test = X_test.drop(['Cover_Type'], axis=1)
X_test
```

```python
clf = GradientBoostingClassifier(max_depth=10, n_estimators=200)
clf.fit(X_train, y_train)

h = clf.predict(X_cv)
print(accuracy_score(y_cv, h))
```

```python
h = clf.predict(X_test)
```

```python
ans = pd.DataFrame({'Id':test_ids, 'Cover_Type': h}, columns=['Id', 'Cover_Type'])
ans.to_csv('data/submit.csv', index=False)
```
